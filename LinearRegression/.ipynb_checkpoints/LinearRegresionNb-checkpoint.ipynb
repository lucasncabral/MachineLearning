#### Linear Regression

Com o intuito de desenvolver o que foi requisitado pelo professor, o vídeo [How to Do Linear Regression using Gradient Descent](https://www.youtube.com/watch?v=XdM6ER7zTLk) forneceu o código python para a regressão linear, com algumas modificações e ajustes, podemos observar o código base que sera utilizado nessa atividade:

```python
# -*- coding: utf-8 -*-
from numpy import *

def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(current_b, current_m, points, learningRate):
    gradient_b = 0
    gradient_m = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        gradient_b += -(2/N) * (y - ((current_m * x) + current_b))
        gradient_m += -(2/N) * x * (y - ((current_m * x) + current_b))
    new_b = current_b - (learningRate * gradient_b)
    new_m = current_m - (learningRate * gradient_m)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

points = genfromtxt("data.csv", delimiter=",")
learning_rate = 0.0001
# y = mx + b
initial_b = 0 
initial_m = 0
num_iterations = 1000

print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
print ("Running...")
[b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))

```

Para realizar o que foi pedido na questão modificamos o metodo step_gradient para coletarmos o gradient_b e gradient_m, com o intuito de no metodo gradient_size criado possamos calcular o tamanho do gradiente. Além disso, no metodo gradient_descent_runner trocas o for que possuiamos por um while, afim de só terminamos a execução quando o nosso algoritmo convergir, que nesse caso foi definido como o tamanho do gradiente ser menor que 0.1.

```python
def step_gradient(current_b, current_m, points, learning_rate):
    gradient_b = 0
    gradient_m = 0
    N = float(len(points))
    
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        
        gradient_b += -(2/N) * (y - ((current_m * x) + current_b))
        gradient_m += -(2/N) * x * (y - ((current_m * x) + current_b))    
    new_b = current_b - (learning_rate * gradient_b)
    new_m = current_m - (learning_rate * gradient_m)
    return [new_b, new_m, gradient_b, gradient_m]

def gradient_size(b,m):
    return math.sqrt(math.pow(b,2) + math.pow(m,2))

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, error):
    b = starting_b
    m = starting_m
        
    iteration = 0
    converged = False
    while(not converged):
        b, m, gradient_b, gradient_m = step_gradient(b, m, array(points), learning_rate)
        gradient_size_values.append(gradient_size(gradient_b, gradient_m))
        iteration += 1
        if(gradient_size(gradient_b, gradient_m) < error):
            converged = True
    return [b, m, iteration]


error = 0.1
gradient_size_values = []  
print ("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
print ("Running...")
[b, m, num_iterations] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, error)
print ("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
```

#### 7. Implemente a forma fechada (equações normais) de calcular os coeficientes de regressão (vide algoritmo nos slides). Compare o tempo de processamento com o gradiente descendente considerando sua solução do item 6.

Metodo que implementa a forma fechada:

```python
def closed_form(points):
    sum_x = 0
    sum_y = 0
    N = float(len(points))
    for i in range(0, len(points)):
        sum_x += points[i, 0]
        sum_y += points[i, 1]
    mean_x = sum_x / N
    mean_y = sum_y / N
    
    erro_x = 0
    superior = 0
    for i in range(0, len(points)):
        x = points[i,0]
        y = points[i,1]
        superior += (x - mean_x) * (y - mean_y)
        erro_x += (x - mean_x) ** 2
    
    m = superior / erro_x
    b = mean_y - m * mean_x
    
    return [b,m]
```

Com o metodo pronto, importamos a biblioteca time afim de calcularmos o tempo de cada método e podermos compara-los.

```python
import time

start_time = time.time()
gradient_descent_runner(points, initial_b, initial_m, learning_rate, error)
end_time = time.time()
print ("Gradiente Descendente - Tempo: "  + str(end_time - start_time))

start_time = time.time()
closed_form(points)
end_time = time.time()
print ("Forma Fechada - Tempo: "  + str(end_time - start_time))
```

Podemos observar que o tempo da forma fechada é muito inferior ao tempo do gradiente descendente, sendo assim a forma fechada mais eficiente para a análise em questão.
